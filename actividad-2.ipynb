```python
# Import and pre-visualization
df = pd.read_csv('https://raw.githubusercontent.com/julihdez36/Inteligencia_artificial/main/Forbes%202000%202022.csv',encoding='latin-1',sep=';')
df
```

```python
print(df.shape)
print(df.columns)
```

```python
#depuracion
df.columns = ['Rank','Company','Country','Sales','Profits','Assets','Market_value','desconocido']
df.info()
```

```python
df = df.replace(',','',regex=True)
df = df.replace('\$','',regex=True)
df[:]
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Country'].value_counts().plot(kind='pie', autopct='%1.0f%%')
plt.show()
```

```python
df['Country'].value_counts().plot(kind='bar')
plt.title('segmentacion por pais')
plt.show()
```

```python
df['Country'].value_counts()
```

```python
#histograma de ventas
np.random.seed(8)
x = np.random.normal(df["Sales"])
fig, ax = plt.subplots()
ax.hist(x)
```

```python
df['Country'] = df['Country'].map({"United States":1,"Italy":0})
df['Country'].value_counts()
```

```python
plt.figure(figsize=(8, 6))
plt.title('ventas vs. valor de mercado')
plt.scatter(x= df['Sales'], y =  df['Market_value'], alpha=0.5, c='b', edgecolors='k')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.title('ventas vs. beneficios')
plt.scatter(x= df['Sales'], y =  df['Profits'])
plt.show()
```

```python
f = ['Market_value', 'Profits']
plt.figure(figsize=(8,8))
for i in f:
    sns.barplot(x='Country',y= i,data=df)
    plt.grid()
    plt.show()
```
