```python
#importacion del modelo 
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min
from sklearn import metrics
from scipy.spatial.distance import cdist
from sklearn.datasets import make_blobs
```

```python
help(KMeans())
```

```python
def elbow_method(df, max_clusters=10):
    sum_of_squared_distances = []
    for k in range(3, max_clusters+1):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(df)
        sum_of_squared_distances.append(kmeans.inertia_)
        plt.plot(range(1, max_clusters+1), sum_of_squared_distances, 'bx-')
    plt.xlabel('Número de Clusters (k)')
    plt.ylabel('Suma de las Distancias al Cuadrado')
    plt.title('Curva del Codo')
    plt.show()
```

```python
elbow_method(df)
```

```python
df = df.drop(df[df['Company'].apply(lambda x: not x.isnumeric())].index)
```

```python
df = df.drop(['Company'], axis=1)
```

```python
# How many clusters (k) are required?
# Elbow method

WCSS = [] #"Within-Cluster-Sum-of-Squares"

for i in range(1,11):
    kmeans = KMeans(n_clusters= i,init='k-means++',random_state=0) #pruebas k
    kmeans.fit(df) #entrenamiento
    WCSS.append(kmeans.inertia_) #guardamos resultados
```

```python
print(WCSS)
```

```python
plt.plot(range(1,11),WCSS)
plt.title('The Elbow Point Graph')
plt.xlabel('Number of clusters (K)')
plt.ylabel('WCSS')
plt.show()
```
